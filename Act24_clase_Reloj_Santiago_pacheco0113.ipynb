```python
class Reloj:
    def __init__(self, marca, tipo, correa, resistencia):
        # Atributos iniciales
        self.marca = marca
        self.tipo = tipo
        self.correa = correa
        self.resistencia = resistencia
        self.hora = "00:00"
        self.alarma = "No configurada"

    def ajustarHora(self, nueva_hora):
        self.hora = nueva_hora
        print(f"\n[SISTEMA]: 🕒 {self.marca}: Hora ajustada con éxito a las {self.hora}")

    def configurarAlarma(self, hora_alarma):
        self.alarma = hora_alarma
        print(f"\n[SISTEMA]: ⏰ {self.marca}: Alarma programada para las {self.alarma}")

    def cronometrar(self, segundos):
        print(f"\n[SISTEMA]: ⏱️ {self.marca}: Iniciando cronómetro por {segundos} segundos...")
        print("...3... 2... 1... ¡Tiempo cumplido! 🔔")

# --- Creación de 2 objetos Reloj ---
r1 = Reloj("Rolex", "Automático", "Acero", "100m")
r2 = Reloj("Casio", "Digital", "Resina", "50m")

# --- Menú Interactivo ---
while True:
    print("\n" + "—" * 30)
    print("      MENÚ PRINCIPAL")
    print("—" * 30)
    print(f"1. Usar {r1.marca} ({r1.tipo})")
    print(f"2. Usar {r2.marca} ({r2.tipo})")
    print("3. Salir del programa")

    reloj_op = input("\nElige un reloj (1-3): ")

    if reloj_op == "3":
        print("\nSaliendo del simulador...")
        break

    if reloj_op == "1":
        obj = r1
    elif reloj_op == "2":
        obj = r2
    else:
        print("Opción no válida.")
        continue

    print(f"\n>>> TRABAJANDO CON: {obj.marca} <<<")
    print(f"Atributos: Correa de {obj.correa} | Resistencia: {obj.resistencia}")
    print("a. Ajustar Hora")
    print("b. Configurar Alarma")
    print("c. Cronometrar")

    accion = input("Elige una acción (a, b o c): ").lower()

    if accion == "a":
        h = input("Ingresa la nueva hora (HH:MM): ")
        obj.ajustarHora(h)
    elif accion == "b":
        a = input("¿A qué hora quieres la alarma? (HH:MM): ")
        obj.configurarAlarma(a)
    elif accion == "c":
        s = input("¿Cuántos segundos quieres cronometrar?: ")
        obj.cronometrar(s)
    else:
        print("Acción no reconocida.")

print("  Programa realizado por Santiago Pacheco NL = 45  ")
```
