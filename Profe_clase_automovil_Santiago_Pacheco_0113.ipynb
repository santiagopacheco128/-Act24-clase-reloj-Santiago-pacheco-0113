```python
# Clase automovil, aytributos, marca, placa, precio
# funciones, leerdatos(), mostrardatos()
# crear 3 objetos Carro1, carro2, carro3

class Automovil:
    def __init__(self, marca, placa, precio):
        self.marca = marca
        self.placa = placa
        self.precio = precio

    def leerdatos(self):
        self.marca = input("Ingrese la marca: ")
        self.placa = input("Ingrese la placa: ")
        self.precio = input("Ingrese el precio: ")
        return self.marca, self.placa, self.precio

    def mostrardatos(self):
        print("Marca: ", self.marca)
        print("Placa: ", self.placa)
        print("Precio: ", self.precio)
        return self.marca, self.placa, self.precio

# --- Fin de la definición de la clase ---

# Creando los datos para carro1, carro2, carro3
print("carro numero 1")
carro1 = Automovil("", "", "") # creando el objetos carro1
carro1.leerdatos()
carro1.mostrardatos()
print("*_*_*_*_*_*_*_*_*_*_*_*_*_")

print("carro numero 2")
carro2 = Automovil("", "", "") # creando el objeto carro2
carro2.leerdatos()
carro2.mostrardatos()
print("*_*_*_*_*_*_*_*_*_*_*_*_*_")

print("carro numero 3")
carro3 = Automovil("", "", "") # creando el objeto carro3
carro3.leerdatos()
carro3.mostrardatos()
print("software desarrollado por Santiago Pacheco NL = 43")
```
